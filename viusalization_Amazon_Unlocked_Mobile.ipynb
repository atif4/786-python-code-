```python
#hide earrings 
import warnings
warnings.filterwarnings('ignore')

# EDA pamages 
import pandas as pd
import numpy as np

# Data Viz pakages
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\haier\\Python\\TEXT_DATA_PREPROCESSING\\Data_sets\\text_data_preprocessed_Amazon_Unlocked_Mobile.csv")
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
def get_scatter_plot(c_id1,c_id2):
    # Scatter plot with day against tip
    plt.scatter(df[c_id1], df[c_id2])
    # Adding Title to the Plot
    plt.title("Scatter Plot")
    # Setting the X and Y labels
    plt.xlabel(c_id1)
    plt.ylabel(c_id2)
    plt.xticks(rotation = 90)
    plt.show()
```

```python
get_scatter_plot('Company Name','Price')
```

```python
get_scatter_plot('Company Name','Rating')
```

```python
get_scatter_plot('Company Name', 'Review Votes')
```

```python
def get_scatter_plot2(c_id1,c_id2,c_id3,c_id4):
    # Scatter plot with day against tip
    plt.scatter(df[c_id1], df[c_id2], c = df[c_id3], s = df[c_id4])
    # Adding Title to the Plot
    plt.title("Scatter Plot")
    # Setting the X and Y labels
    plt.xlabel(c_id1)
    plt.ylabel(c_id2)
    plt.xticks(rotation = 90)
    plt.colorbar()
    plt.show()
```

```python
get_scatter_plot2('Company Name', 'Price', 'Rating', 'Review Votes')
```

```python
def get_line_plot(c_id1,c_id2):
    # Scatter plot with day against tip
    plt.plot(df[c_id1])
    plt.plot(df[c_id2])
    # Adding Title to the Plot
    plt.title("Line Plot")
    # Setting the X and Y labels
    plt.xlabel(c_id1)
    plt.ylabel(c_id2)
    #plt.xticks(rotation = 90)
    plt.show()
```

```python
get_line_plot('Company Name','Price')
```

```python
get_line_plot('Company Name','Rating')
```

```python
get_line_plot('Company Name', 'Review Votes')
```

```python
def get_his_plot(c_id1):
    # Scatter plot with day against tip
    plt.hist(df[c_id1])
    # Adding Title to the Plot
    plt.title("Histogram")
    #plt.xticks(rotation = 90)
    plt.show()# reading the database
```
